```python
import requests as req

from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
```

```python
#most streamed spotify artists
url = 'https://chartmasters.org/most-streamed-artists-ever-on-spotify/'
```

```python
html = req.get(url).content
soup = bs(html, 'html.parser')
```

```python
artist_spoti = pd.read_html(html)[0]
artist_spoti.sample()
```

```python
artist_spoti.rename(columns={"#":"streaming_rank",'Artist Name': 'artist', "Lead Streams":"lead_streams", "Featured Streams":"featured_streams", "Last Update":"last_update"}, inplace=True)
```

```python
artist_spoti.dropna(subset=["artist"], inplace = True)
artist_spoti.shape
```

```python
artist_spoti.drop_duplicates(subset=["artist"], inplace=True)
```
